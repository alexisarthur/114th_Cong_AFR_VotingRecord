Definitions: 

    - Congress Number
    - ICPSR ID Number:  5 digit code assigned by the ICPSR 
    - ICPSR ID Number:  original 5 digit ICPSR code (this field is usually blank).
    - State Code:  2 digit ICPSR State Code.
    - Congressional District Number (0 if Senate)
    - State Name: 
    - Party Code:  100 = Dem., 200 = Repub. (See PARTY3.DAT)
    - ICPSR Occupancy Code:  See any ICPSR Roll Call Voting Codebook
    - ICPSR Office Code:  See any ICPSR Roll Call Voting Codebook
    - 1st Dimension Coordinate
    - 2nd Dimension Coordinate
  
On members specifically: 
    - 1st Dimension Coordinate
    - 2nd Dimension Coordinate
    - 1st Dimension Conditional Standard Error
    - 2nd Dimension Conditional Standard Error
    - Log-Likelihood
    - Number of Votes
    - Number of Classification Errors
    - Geometric Mean Probability
    - last_means (How they attained office appointed, elected, etc)
    
    
Voting Key: 
  0=not a member
  1=Yea
  2=Paired Yea
  3=Announced Yea
  4=Announced Nay 
  5=Paired Nay
  6=Nay,
  7=Present (some Congresses, also not used some Congresses),
  8=Present (some Congresses, also not used some Congresses),
  9=Not Voting    
  
  For reference: https://legacy.voteview.com/house113.htm

```python
headers[1] = 'chamber'
headers[7] = 'state'
headers[16] = 'clerk_rollnumber'
headers[17] = 'date'
headers[18] = 'vote_desc'
headers[19] = 'party'

df.columns = headers
df.head(3)
```

```python
df.vote_result.unique()
```

```python
def vote_result(x):
    if x== 'Passed':
        return 1
    elif x== 'Failed':
        return 0
    elif x== 'Agreed to':
        return 2
    elif x== 'Ryan (WI)':
        return 1
    elif x== 'Boehner':
        return 1
    else:
        return 'NA'
```

```python
df['vote_result_val'] = df['vote_result'].apply(vote_result)

df
```

```python
def castcode(x):
    if x== 1:
        return 'Yea'
    elif x== 2:
        return 'Paired Yea'
    elif x== 3:
        return 'Announced Yea'
    elif x== 4:
        return 'Announced Nay'
    elif x== 5:
        return 'Paired Nay'
    elif x== 6:
        return 'Nay'
    elif x== 9:
        return 'Not Voting'
    else:
        return 'NA'
```

```python
df.cast_code.replace(to_replace=6,value=0, inplace=True)
```

```python
df.cast_code.replace(9, np.nan,inplace=True)
```

```python
df.cast_code.unique()
```

```python
df['cast_desc'] = df['cast_code'].apply(castcode)

df
```

df.to_csv('AFR_analysis.csv',header=True, index=False)
